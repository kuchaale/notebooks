```python
#- Dataset
database = "/Users/lesommer/data/NATL60/NATL60-MJM155-S/1d/2008/"
files_march = sorted(glob(database + 'NATL60-MJM155_y2008m03d*.1d_CURLOVERF3D.nc'))
files_sep = sorted(glob(database + 'NATL60-MJM155_y2008m09d*.1d_CURLOVERF3D.nc'))
```

```python
socurl0 = Dataset(files_march[0]).variables['socurloverf']
asocurl0 = socurl0[...,2000:,1500:].squeeze()
```

```python
from pyClim.filTools import lanczos
n = 30
fc = 1./120 # cycle per observation
lowpass = lanczos('lp', [n,n],fc1=[fc, fc])
```

```python
%time lpsocurl = lowpass.do_filtering2d(asocurl0)
```

```python
# from  https://anaconda.org/dask/parallelize_image_filtering_workload/notebook
import dask.array as da
tab = asocurl0
mask = asocurl0.mask
```

```python
from multiprocessing import cpu_count
cpu_count()
```

```python
chunk_size = tuple([x//2 for x in tab.shape])
tab.shape,chunk_size
```

```python
filtering = lambda t : lowpass.do_filtering2d(t)
tab_da = da.from_array(np.array(tab),chunks=chunk_size)
%time  lpsocurl_da = tab_da.map_overlap(filtering, depth=n).compute()
```

```python
filtering = lambda t : lowpass.do_filtering2d(t)
tab_da = da.from_array(np.array(tab),chunks=chunk_size)
%time  lpsocurl_da = tab_da.map_blocks(filtering).compute()
```

```python
filtering = lambda t : lowpass.do_filtering2d(t,fast=True)
tab_da = da.from_array(np.array(tab),chunks=chunk_size)
%time  lpsocurl_da = tab_da.map_blocks(filtering).compute()
```

```python
def nice_plot(lpsocurl,asocurl0):
    fig, axarr = plt.subplots(1,3)
    fig.set_figheight(4)
    fig.set_figwidth(18)
    
    axarr[0].pcolormesh(asocurl0,cmap=div_cmap,vmin=-1,vmax=1)
    axarr[0].set_title('Full',fontsize=20)
    axarr[0].axis('tight')

    axarr[1].pcolormesh(lpsocurl,cmap=div_cmap,vmin=-1,vmax=1)
    axarr[1].set_title('Low passed',fontsize=20)
    axarr[1].axis('tight')

    axarr[2].pcolormesh(asocurl0 - lpsocurl,cmap=div_cmap,vmin=-1,vmax=1)
    axarr[2].set_title('Difference',fontsize=20)
    axarr[2].axis('tight')

nice_plot(lpsocurl_da,asocurl0)
```
